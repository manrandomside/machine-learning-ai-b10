## Fashion MNIST Image Classification with TensorFlow/Keras

> Tugas ini bertujuan untuk membangun model Multi-Layer Perceptron Sederhana untuk klasifikasi pakaian dari dataset "Fashion MNIST"

> Dataset berasal dari ```fashion_mnist``` yang tersedia di ```tensorflow.keras.datasets```

> Tugas ini tidak memiliki referensi, silakan bereksperimen dibawah ini :)

### 1. Import module yang diperlukan

````python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from sklearn.metrics import classification_report, confusion_matrix

# Set seed untuk reproducibility
np.random.seed(42)
tf.random.set_seed(42)
````

### 2. Data preprocess

````python
# 2. Load Dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalisasi
X_train = X_train / 255.0
X_test = X_test / 255.0
````

### 3.  Data visualization

````python
# Nama kelas Fashion MNIST
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plt.figure(figsize=(12, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"{class_names[y_train[i]]}")
    plt.axis('off')
plt.suptitle("Sample Gambar dari Fashion MNIST")
plt.tight_layout()
plt.show()
````

### 4. Pembangunan Model

### TODO #1 :

Untuk model MLP\
**NOTE! : Jalankan pertama kali blok 4 ini dulu, baru cobain sesuai yang sudah tertulis dibawah setelah dapat hasilnya :**

DAFTAR HYPERPARAMETER
- Coba lakukan tuning hyperparameter pada beberapa parameter di model MLP ini.
- Ganti ukuran hidden layer (saat ini : ```[128, 64, 10]```, coba perbesar atau perkecil).
- Coba tambah 1-2 layer lagi.
- Naikkan epoch ke ```20```, ```30``` atau ```50```.
- Ganti activation menjadi selain relu untuk layer Dense.
- Ganti optimizer ke yang lain selain ```adam```, coba ```sgd``` atau ```rmsprop```.

````python
# Model MLP Default (Baseline)
model_mlp = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
model_mlp.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
model_mlp.summary()
````

### 4a. Pembangunan Model MLP (Tuned)

> Setelah model MLP default dijalankan, di sini dilakukan tuning hyperparameter dengan mengubah:
>
> - **Tambah 1 hidden layer** : jadi total 4 hidden layers untuk meningkatkan kapasitas model
> - **Perbesar ukuran neuron** : dari [128, 64] menjadi [256, 128, 64, 32]
> - **Ganti optimizer** : dari adam ke rmsprop
> - **Naikkan epoch** : dari 10 ke 20 (di section training)

````python
# Model MLP Tuned (setelah hyperparameter tuning)
model_mlp_tuned = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(10, activation='softmax')
])
model_mlp_tuned.compile(optimizer='rmsprop',
                        loss='sparse_categorical_crossentropy',
                        metrics=['accuracy'])
model_mlp_tuned.summary()
````

### 4b. Pembangunan Model CNN

### TODO #2 :

Untuk model CNN
- Komentar kode di blok 4 Pembangunan Model MLP
- Jalankan blok dibawah.
- Coba training tanpa lakukan hyperparameter tuning apapun
- Berikan insight dibawah

````python
# Model CNN (tanpa tuning)
model_cnn = models.Sequential([
    layers.Reshape((28, 28, 1), input_shape=(28, 28)),
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
model_cnn.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
model_cnn.summary()
````

### 5. Training

````python
# Training MLP Default
print("=" * 50)
print("Training MLP Default")
print("=" * 50)
history_mlp = model_mlp.fit(X_train, y_train, epochs=10, validation_split=0.2, verbose=2)
````

````python
# Training MLP Tuned
print("=" * 50)
print("Training MLP Tuned")
print("=" * 50)
history_mlp_tuned = model_mlp_tuned.fit(X_train, y_train, epochs=20, validation_split=0.2, verbose=2)
````

````python
# Training CNN
print("=" * 50)
print("Training CNN")
print("=" * 50)
history_cnn = model_cnn.fit(X_train, y_train, epochs=10, validation_split=0.2, verbose=2)
````

### 6. Evaluasi & Visualisasi

````python
# Evaluasi semua model di data test
loss_mlp, acc_mlp = model_mlp.evaluate(X_test, y_test, verbose=0)
loss_mlp_tuned, acc_mlp_tuned = model_mlp_tuned.evaluate(X_test, y_test, verbose=0)
loss_cnn, acc_cnn = model_cnn.evaluate(X_test, y_test, verbose=0)

print(f"MLP Default     | Loss: {loss_mlp:.4f} | Akurasi: {acc_mlp:.4f}")
print(f"MLP Tuned       | Loss: {loss_mlp_tuned:.4f} | Akurasi: {acc_mlp_tuned:.4f}")
print(f"CNN             | Loss: {loss_cnn:.4f} | Akurasi: {acc_cnn:.4f}")
````

````python
# Plot perbandingan akurasi training & validation
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history_mlp.history['val_accuracy'], label='MLP Default', linewidth=2)
axes[0].plot(history_mlp_tuned.history['val_accuracy'], label='MLP Tuned', linewidth=2)
axes[0].plot(history_cnn.history['val_accuracy'], label='CNN', linewidth=2)
axes[0].set_title('Validation Accuracy per Epoch')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(history_mlp.history['val_loss'], label='MLP Default', linewidth=2)
axes[1].plot(history_mlp_tuned.history['val_loss'], label='MLP Tuned', linewidth=2)
axes[1].plot(history_cnn.history['val_loss'], label='CNN', linewidth=2)
axes[1].set_title('Validation Loss per Epoch')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

### 7. Evaluasi lanjutan

````python
# Evaluasi lanjutan untuk model terbaik (CNN)
print("=" * 60)
print("CLASSIFICATION REPORT - CNN Model")
print("=" * 60)

y_pred_cnn = model_cnn.predict(X_test, verbose=0)
y_pred_classes_cnn = np.argmax(y_pred_cnn, axis=1)

print(classification_report(y_test, y_pred_classes_cnn, target_names=class_names))
````

````python
import seaborn as sns

cm = confusion_matrix(y_test, y_pred_classes_cnn)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('Confusion Matrix - CNN Model')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
````
