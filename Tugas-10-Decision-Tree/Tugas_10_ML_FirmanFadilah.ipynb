### 2. Data collection

````python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Optional: map target to actual class names
df['species'] = df['target'].map(dict(enumerate(iris.target_names)))
df.head()
````

### TODO #1 :

> Isi insight, coba cari tahu apa itu "sepal length, sepal width, petal length, petal width" pada bunga iris.

> Bunga iris punya dua jenis "daun bunga" yang dipakai sebagai fitur:
>
> - **Sepal** itu kelopak luar bunga, yang biasanya berwarna hijau dan fungsinya melindungi bunga saat masih kuncup. Pada iris, sepal-nya cukup besar dan menyerupai daun mahkota.
> - **Petal** itu mahkota bunga, bagian yang berwarna cerah (ungu, putih, kuning) yang biasanya menarik perhatian.
>
> Lalu untuk **length** (panjang) dan **width** (lebar):
>
> - **Sepal length** : panjang kelopak luar (dari pangkal sampai ujung)
> - **Sepal width** : lebar kelopak luar (di bagian terlebar)
> - **Petal length** : panjang mahkota bunga
> - **Petal width** : lebar mahkota bunga
>
> Keempat ukuran ini diukur dalam satuan sentimeter. Yang menarik, dari pengamatan botanis sebelumnya, ukuran petal (mahkota) ternyata jauh lebih membedakan spesies iris dibanding sepal. Setosa punya petal yang sangat kecil, sementara Virginica punya petal yang paling besar. Karena perbedaan inilah dataset ini bisa diprediksi dengan akurasi tinggi pakai model machine learning sederhana sekalipun.
